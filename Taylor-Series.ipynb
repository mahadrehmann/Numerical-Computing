```python
# Importing libraries 
import matplotlib.pyplot as plt 
import numpy as np 
import math 

# Using Numpy to create an array X 
X = np.arange(0, 3/2*math.pi, 0.05) 

term1 = [np.cos(1)] * len(X)
term2 = -np.sin(1)*(X-1)
term3 = -1/2*np.cos(1)*(X-1)**2
term4 = 1/6*np.sin(1)*(X-1)**3
term5 = 1/24*np.cos(1)*(X-1)**4

# Assign variables to the y axis part of the curve 
y = np.cos(X) 
y1 = term1 + term2 + term3 + term4 + term5

# Plotting both the curves simultaneously 
plt.plot(X, y, color='r', label='cos') 
plt.plot(X, y1, color='g', label='cos approx') 

# Naming the x-axis, y-axis and the whole graph 
plt.xlabel("Angle") 
plt.ylabel("Magnitude") 
plt.title("Tayler terms of Cosine function") 

# Adding legend, which helps us recognize the curve according to it's color 
plt.legend() 

# To load the display window 
plt.show() 
```
